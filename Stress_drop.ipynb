```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
```

```python
filename = 'stress_drop.pd'

data = pd.read_csv(filename, delim_whitespace=True)
data.head(10)
```

```python
N  = data['ID'].max()
SD = np.zeros((N,1))

for id in range(1,N+1):
    mean_stress  = data[(data['ID'] == id) & (data['R2'] >= 80) & (data['SNR']>5) & (data['fcut'] <=19 )]['Stress' ].mean()
    std_stress   = data[(data['ID'] == id) & (data['R2'] >= 80) & (data['SNR']>5) & (data['fcut'] <=19 )]['Stress' ].std()
    SD[id-1]       = mean_stress
    print(id ,'-', '%5.2f'%SD[id-1][0])
    #print('id: ',id, ' str: ', mean_stress, ' (+/-) ', std_stress)
#for id in range(1,N+1):

np.savetxt('stress.out', SD, fmt='%5.2f') 
```
